```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 1000)
```

```python
barcelona = pd.read_csv('../datasets/barcelona/barcelona_final.csv')
barcelona
```

```python
def rate_change(x):
    x = list(x)
    if len(x) > 1:
        aggregator = 0
        for i in range(len(x)-1):
            aggregator += ((x[i+1] - x[i])/x[i])*100
        return aggregator/len(x)
    else: return np.nan

rows_num = barcelona['room_id'].value_counts()
multi_price_rooms = barcelona[barcelona['room_id'].isin(rows_num[rows_num > 1].index)]
price_change = multi_price_rooms.groupby(['neighborhood', 'room_id'])['price'].agg(rate_change)
price_change
```

```python
no_outlrs_brc = pd.DataFrame(barcelona)

for percentage in price_change:
    if percentage > 200:
        room_id = price_change[price_change == percentage].index[0][1]
        room_with_outlr = no_outlrs_brc[no_outlrs_brc['room_id'] == room_id]
        row_with_outlr = room_with_outlr[room_with_outlr['price'] == room_with_outlr['price'].max()]
        no_outlrs_brc.drop(row_with_outlr.index, inplace=True)
    else: continue
```
