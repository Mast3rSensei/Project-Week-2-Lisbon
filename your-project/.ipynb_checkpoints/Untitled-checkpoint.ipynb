```python
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_rows', 1000)
```

```python
brc_files = os.listdir('../datasets/barcelona')
brc_files
```

```python
barcelona = pd.concat([pd.read_csv('../datasets/barcelona/' + brc_files[i]) for i in range(0, 13)], axis = 0, ignore_index = True)
barcelona2 = pd.concat([pd.read_csv('../datasets/barcelona/' + brc_files[i]) for i in range(13, len(brc_files))], axis = 0, ignore_index = True)
```

```python
barcelona = pd.concat([barcelona, barcelona2[barcelona.columns]], ignore_index=True)
barcelona = barcelona.sort_values(['room_id'])
barcelona.head(100)
```

```python
barcelona['minstay'].isna().count()
```

```python
def rate_change(x):
    x = list(x)
    if len(x) > 1:
        aggregator = 0
        for i in range(len(x)-1):
            aggregator += ((x[i+1] - x[i])/x[i])*100
        return aggregator/len(x)
    else: return np.nan

price_change = barcelona.groupby('room_id')['price'].agg(rate_change)
```

```python
price_change.dropna(inplace=True)
price_change.mean()
```

```python
no_outlrs_brc = pd.DataFrame(barcelona)

for percentage in price_change:
    if percentage > 200:
        room_id = price_change[price_change == percentage].index[0]
        room_with_outlr = no_outlrs_brc[no_outlrs_brc['room_id'] == room_id]
        row_with_outlr = room_with_outlr[room_with_outlr['price'] == room_with_outlr['price'].max()]
        no_outlrs_brc.drop(row_with_outlr.index, inplace=True)
```

```python
no_outlrs_brc.head(100)
```

```python
price_change2 = no_outlrs_brc.groupby('room_id')['price'].agg(rate_change)
price_change2.dropna(inplace = True)
price_change2.mean()
```
