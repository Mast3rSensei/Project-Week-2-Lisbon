```python
brc_files = os.listdir('../datasets/barcelona')
brc_files
```

```python
barcelona = pd.concat([pd.read_csv('../datasets/barcelona/' + brc_files[i]) for i in range(0, 13)], axis = 0, ignore_index = True)
barcelona
```

```python
barcelona = barcelona.sort_values(['room_id', 'last_modified'])
barcelona.head(50)
```

```python
def rate_change(x):
    x = list(x)
    if len(x) > 1:
        aggregator = 0
        for i in range(len(x)-1):
            aggregator += (x[i+1] - x[i])
        return aggregator/len(x)
    else: return np.nan

#sr = barcelona['price'][barcelona['room_id'] == 10938]
#print(list(sr))
#rate_change(list(sr))
price_change = barcelona.groupby('room_id')['price'].agg(rate_change)
```
